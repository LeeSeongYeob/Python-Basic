# Matplotlib Graph 2
- scatter
- bar
- Histogram

```
import matplotlib.pyplot as plt
import numpy as np
```

```
fig, ax = plt.subplots()
x = np.arange(10)
ax.plot(
    x,x**2,'o',
    markersize=10,
    markerfacecolor='white',
    markeredgecolor = 'blue'
)
fig.show()
```

```
fig, ax = plt.subplots()
x = np.random.randn(50)
y = np.random.randn(50)
colors = np.random.randint(0, 100,
50)
sizes = 500 * np.pi * np.random.rand(50) ** 2
ax.scatter(
x, y, c=colors, s=sizes,
alpha=0.3
)
fig.show()
```

```
x = np.arange(10)
fig, ax = plt.subplots(figsize = (12,4))
ax.bar(x,x**2)
fig.show()
```

```
x = np.random.rand(3)
y = np.random.rand(3)
z = np.random.rand(3)
data = [x, y, z]

fig, ax = plt.subplots()
x_ax = np.arange(3)
for i in x_ax:
    ax.bar(x_ax, data[i],
           bottom=np.sum(data[:i],axis=0))
ax.set_xticks(x_ax)
ax.set_xticklabels(['A','B','C'])
```

```
fig, ax = plt.subplots()
data = np.random.randn(1000)
ax.hist(data, bins=50)
fig.show()
```
