# Matplotlib Graph 그리기.
- figure 생성
- plot 그리기

```
import matplotlib.pyplot as plt
import numpy as np
```

##### plot로 그래프 그리기

```
x = [1,2,3,4,5]
y = [1,2,3,4,5]
plt.plot(x,y)
```

```
fig, ax = plt.subplots()
ax.plot(x,y)
ax.set_title('First plot')
ax.set_xlabel('x')
ax.set_ylabel('y')
```

##### 1 figure 에 2개 그래프 그리기'

```
x = np.linspace(0,np.pi*4,100)
fig, ax = plt.subplots(2, 1)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```
fig, ax = plt.subplots()
x = np.arange(15)
y = x ** 2
ax.plot(
    x,y,
    linestyle=':',
    marker='*',
    color = 'm'
)
```

##### 축 경계 조정하기

```
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()
ax.plot(x, np.sin(x))
ax.set_xlim(-2,12)
ax.set_ylim(-1.5,1.5)
```

##### legend 범례 추가

```
fig, ax = plt.subplots()
ax.plot(x, x, label='y=x')
ax.plot(x, x**2, label = 'y=x^2')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.legend(
    loc = 'upper right',
    shadow = True,
    fancybox = True,
    borderpad = 2
)
```
